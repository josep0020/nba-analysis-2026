# NBA Player Statistics Analysis 2026

## Project Overview

This project performs exploratory data analysis on NBA player statistics for the 2026 season. The goal is to identify the league's top performers across multiple categories: scoring, efficiency, playmaking, and defense.

### Objectives:
- **Scoring Leaders**: Identify top point scorers
- **Efficiency Analysis**: Evaluate field goal percentages and efficiency ratings
- **Playmaking**: Analyze assists and assist-to-turnover ratios
- **Defense**: Examine rebounds, steals, and blocks
- **Team Analysis**: Compare team performance metrics

---

```python
import pandas as pd
import numpy as np
import warnings
warnings.filterwarnings('ignore')
```

```python
# Load the dataset
nba = pd.read_csv('nba_player_stats_2026.csv')

print(f"Dataset loaded successfully!")
print(f"Shape: {nba.shape[0]} players, {nba.shape[1]} features")
```

## 1. Data Exploration

```python
# Display columns and first row
print("Columns in the dataset:")
print(nba.columns.tolist())
print("\nSample data:")
nba.head(1)
```

```python
# Basic statistics
print("--- Dataset Statistics ---")
print(f"Total Players: {nba.shape[0]}")
print(f"Total Teams: {nba['TEAM'].nunique()}")
print(f"Missing Values: {nba.isnull().sum().sum()}")
print(f"\nAverage Points Per Player: {nba['PTS'].mean():.2f}")
print(f"Average Efficiency: {nba['EFF'].mean():.2f}")
```

## 2. Top Scorers (Points Leaders)

```python
# Top 5 scorers with points per game
scorers = nba.nlargest(10, 'PTS')[['PLAYER', 'TEAM', 'GP', 'PTS', 'FG_PCT', 'EFF']].copy()
scorers['PPG'] = (scorers['PTS'] / scorers['GP']).round(2)
scorers['FG_PCT'] = (scorers['FG_PCT'] * 100).round(1).astype(str) + '%'

print("=== TOP 10 SCORERS ===")
scorers.head(10)
```

## 3. Efficiency Analysis

```python
# Most efficient players (by EFF rating)
efficient = nba.nlargest(10, 'EFF')[['PLAYER', 'TEAM', 'EFF', 'FG_PCT', 'PTS']].copy()
efficient['FG_PCT'] = (efficient['FG_PCT'] * 100).round(1).astype(str) + '%'

print("=== TOP 10 BY EFFICIENCY (EFF) ===")
efficient
```

```python
# Best field goal percentage (min 200 FGA for relevance)
fg_min = nba[nba['FGA'] >= 200].nlargest(10, 'FG_PCT')[['PLAYER', 'TEAM', 'FG_PCT', 'FGM', 'FGA']].copy()
fg_min['FG_PCT'] = (fg_min['FG_PCT'] * 100).round(1).astype(str) + '%'

print("=== TOP 10 FIELD GOAL % (min 200 FGA) ===")
fg_min
```

## 4. Playmaking (Assists)

```python
# Top assist leaders
assists = nba.nlargest(10, 'AST')[['PLAYER', 'TEAM', 'AST', 'AST_TOV']].copy()
assists['APG'] = (assists['AST'] / nba.loc[assists.index, 'GP']).round(2)

print("=== TOP 10 ASSIST LEADERS ===")
assists
```

```python
# Best assist-to-turnover ratio (min 100 assists)
tov_ratio = nba[nba['AST'] >= 100].nlargest(10, 'AST_TOV')[['PLAYER', 'TEAM', 'AST_TOV', 'AST', 'TOV']].copy()

print("=== BEST ASSIST-TO-TURNOVER RATIO (min 100 assists) ===")
tov_ratio
```

## 5. Rebounding & Defense

```python
# Top rebounders
rebounds = nba.nlargest(10, 'REB')[['PLAYER', 'TEAM', 'REB', 'OREB', 'DREB']].copy()
rebounds['RPG'] = (rebounds['REB'] / nba.loc[rebounds.index, 'GP']).round(2)

print("=== TOP 10 REBOUNDERS ===")
rebounds
```

```python
# Top blockers
blocks = nba.nlargest(10, 'BLK')[['PLAYER', 'TEAM', 'BLK']].copy()
blocks['BPG'] = (blocks['BLK'] / nba.loc[blocks.index, 'GP']).round(2)

print("=== TOP 10 BLOCKERS ===")
blocks
```

```python
# Top thieves (steals)
steals = nba.nlargest(10, 'STL')[['PLAYER', 'TEAM', 'STL']].copy()
steals['SPG'] = (steals['STL'] / nba.loc[steals.index, 'GP']).round(2)

print("=== TOP 10 STEAL LEADERS ===")
steals
```

## 6. Three-Point Shooting

```python
# Best 3-point shooters (min 150 attempts)
three_pt = nba[nba['FG3A'] >= 150].nlargest(10, 'FG3_PCT')[['PLAYER', 'TEAM', 'FG3_PCT', 'FG3M', 'FG3A']].copy()
three_pt['FG3_PCT'] = (three_pt['FG3_PCT'] * 100).round(1).astype(str) + '%'

print("=== TOP 10 THREE-POINT % (min 150 attempts) ===")
three_pt
```

## 7. Team Analysis

```python
# Team scoring totals
team_stats = nba.groupby('TEAM').agg({
    'PTS': ['sum', 'mean'],
    'EFF': 'sum',
    'PLAYER': 'count'
}).reset_index()
team_stats.columns = ['TEAM', 'TOTAL_PTS', 'AVG_PTS', 'TOTAL_EFF', 'PLAYER_COUNT']
team_stats = team_stats.sort_values('TOTAL_PTS', ascending=False)

print("=== TEAM SCORING TOTALS ===")
team_stats.head(10)
```

## 8. Key Insights & Conclusions

```python
# Summary insights
print("══════════════════════════════════════════════════════")
print("                    KEY INSIGHTS")
print("══════════════════════════════════════════════════════")

top_scorer = nba.loc[nba['PTS'].idxmax()]
top_eff = nba.loc[nba['EFF'].idxmax()]
top_ast = nba.loc[nba['AST'].idxmax()]
top_blk = nba.loc[nba['BLK'].idxmax()]
top_stl = nba.loc[nba['STL'].idxmax()]

print(f"\n🏀 SCORING:")
print(f"   • {top_scorer['PLAYER']} leads with {top_scorer['PTS']} points ({top_scorer['PTS']/top_scorer['GP']:.1f} PPG)")

print(f"\n⚡ EFFICIENCY:")
print(f"   • {top_eff['PLAYER']} has the highest efficiency ({top_eff['EFF']})")

print(f"\n🎯 PLAYMAKING:")
print(f"   • {top_ast['PLAYER']}: {top_ast['AST']} assists ({top_ast['AST']/top_ast['GP']:.1f} APG)")

print(f"\n🛡️ DEFENSE:")
print(f"   • {top_blk['PLAYER']} leads blocks with {top_blk['BLK']}")
print(f"   • {top_stl['PLAYER']} leads steals with {top_stl['STL']}")

print(f"\n📊 TEAM ANALYSIS:")
print(f"   • {team_stats.iloc[0]['TEAM']} has highest total points: {team_stats.iloc[0]['TOTAL_PTS']}")
print(f"   • {nba['TEAM'].nunique()} teams represented in the league")

print("\n══════════════════════════════════════════════════════")
```
