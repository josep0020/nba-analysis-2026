# Análisis Exploratorio de Jugadores NBA 2026

## Resumen del Proyecto

Este proyecto realiza un análisis exploratorio de datos (EDA) sobre las estadísticas de los jugadores de la NBA para la temporada 2026. El objetivo es identificar a los mejores jugadores de la liga en múltiples categorías: anotación, eficiencia, creación de juego y defensa.

### Objetivos:
- **Líderes de Anotación**: Identificar a los máximos anotadores.
- **Análisis de Eficiencia**: Evaluar porcentajes de tiro y ratings de eficiencia.
- **Creación de Juego**: Analizar asistencias y ratios de asistencia/pérdida.
- **Defensa**: Examinar rebotes, robos y tapones.
- **Análisis por Equipo**: Comparar métricas de rendimiento por equipo.

---

```python
import pandas as pd
import numpy as np
import warnings
warnings.filterwarnings('ignore')
```

```python
# Cargar el dataset
nba = pd.read_csv('nba_player_stats_2026.csv')

print(f"¡Dataset cargado con éxito!")
print(f"Forma: {nba.shape[0]} jugadores, {nba.shape[1]} columnas")
```

## 1. Exploración de Datos

```python
# Mostrar columnas y primera fila
print("Columnas en el dataset:")
print(nba.columns.tolist())
print("\nMuestra de datos:")
nba.head(1)
```

```python
# Estadísticas básicas
print("--- Estadísticas del Dataset ---")
print(f"Total de Jugadores: {nba.shape[0]}")
print(f"Total de Equipos: {nba['TEAM'].nunique()}")
print(f"Valores Faltantes: {nba.isnull().sum().sum()}")
print(f"\nPromedio de Puntos por Jugador: {nba['PTS'].mean():.2f}")
print(f"Eficiencia Promedio: {nba['EFF'].mean():.2f}")
```

## 2. Máximos Anotadores

```python
# Top 10 anotadores con puntos por partido (PPG)
anotadores = nba.nlargest(10, 'PTS')[['PLAYER', 'TEAM', 'GP', 'PTS', 'FG_PCT', 'EFF']].copy()
anotadores['PPG'] = (anotadores['PTS'] / anotadores['GP']).round(2)
anotadores['FG_PCT'] = (anotadores['FG_PCT'] * 100).round(1).astype(str) + '%'

print("=== TOP 10 ANOTADORES ===")
anotadores
```

## 3. Análisis de Eficiencia

```python
# Jugadores más eficientes (por rating EFF)
eficientes = nba.nlargest(10, 'EFF')[['PLAYER', 'TEAM', 'EFF', 'FG_PCT', 'PTS']].copy()
eficientes['FG_PCT'] = (eficientes['FG_PCT'] * 100).round(1).astype(str) + '%'

print("=== TOP 10 POR EFICIENCIA (EFF) ===")
eficientes
```

```python
# Mejor porcentaje de tiros de campo (mínimo 200 FGA para relevancia)
fg_min = nba[nba['FGA'] >= 200].nlargest(10, 'FG_PCT')[['PLAYER', 'TEAM', 'FG_PCT', 'FGM', 'FGA']].copy()
fg_min['FG_PCT'] = (fg_min['FG_PCT'] * 100).round(1).astype(str) + '%'

print("=== TOP 10 % TIROS DE CAMPO (mín. 200 FGA) ===")
fg_min
```

## 4. Creación de Juego (Asistencias)

```python
# Líderes en asistencias
asistencias = nba.nlargest(10, 'AST')[['PLAYER', 'TEAM', 'AST', 'AST_TOV']].copy()
asistencias['APG'] = (asistencias['AST'] / nba.loc[asistencias.index, 'GP']).round(2)

print("=== TOP 10 LÍDERES EN ASISTENCIAS ===")
asistencias
```

```python
# Mejor ratio asistencia-pérdida (mínimo 100 asistencias)
ratio_ast_tov = nba[nba['AST'] >= 100].nlargest(10, 'AST_TOV')[['PLAYER', 'TEAM', 'AST_TOV', 'AST', 'TOV']].copy()

print("=== MEJOR RATIO ASISTENCIA/PÉRDIDA (mín. 100 asistencias) ===")
ratio_ast_tov
```

## 5. Rebotes y Defensa

```python
# Mejores reboteadores
rebotes = nba.nlargest(10, 'REB')[['PLAYER', 'TEAM', 'REB', 'OREB', 'DREB']].copy()
rebotes['RPG'] = (rebotes['REB'] / nba.loc[rebotes.index, 'GP']).round(2)

print("=== TOP 10 REBOTEADORES ===")
rebotes
```

```python
# Máximos taponadores
tapones = nba.nlargest(10, 'BLK')[['PLAYER', 'TEAM', 'BLK']].copy()
tapones['BPG'] = (tapones['BLK'] / nba.loc[tapones.index, 'GP']).round(2)

print("=== TOP 10 TAPONADORES ===")
tapones
```

```python
# Líderes en robos
robos = nba.nlargest(10, 'STL')[['PLAYER', 'TEAM', 'STL']].copy()
robos['SPG'] = (robos['STL'] / nba.loc[robos.index, 'GP']).round(2)

print("=== TOP 10 LÍDERES EN ROBOS ===")
robos
```

## 6. Tiros de Tres Puntos

```python
# Mejores tiradores de triples (mínimo 150 intentos)
triples = nba[nba['FG3A'] >= 150].nlargest(10, 'FG3_PCT')[['PLAYER', 'TEAM', 'FG3_PCT', 'FG3M', 'FG3A']].copy()
triples['FG3_PCT'] = (triples['FG3_PCT'] * 100).round(1).astype(str) + '%'

print("=== TOP 10 % TRIPLES (mín. 150 intentos) ===")
triples
```

## 7. Análisis por Equipo

```python
# Totales de anotación por equipo
stats_equipo = nba.groupby('TEAM').agg({
    'PTS': ['sum', 'mean'],
    'EFF': 'sum',
    'PLAYER': 'count'
}).reset_index()
stats_equipo.columns = ['EQUIPO', 'TOTAL_PTS', 'PROMEDIO_PTS', 'TOTAL_EFF', 'CANTIDAD_JUGADORES']
stats_equipo = stats_equipo.sort_values('TOTAL_PTS', ascending=False)

print("=== TOTALES DE ANOTACIÓN POR EQUIPO ===")
stats_equipo.head(10)
```

## 8. Conclusiones y Hallazgos Clave

```python
# Resumen de hallazgos
print("══════════════════════════════════════════════════════")
print("                    HALLAZGOS CLAVE")
print("══════════════════════════════════════════════════════")

top_anotador = nba.loc[nba['PTS'].idxmax()]
top_eficiencia = nba.loc[nba['EFF'].idxmax()]
top_asistencias = nba.loc[nba['AST'].idxmax()]
top_tapones = nba.loc[nba['BLK'].idxmax()]
top_robos = nba.loc[nba['STL'].idxmax()]

print(f"\n ANOTACIÓN:")
print(f"   • {top_anotador['PLAYER']} lidera con {top_anotador['PTS']} puntos ({top_anotador['PTS']/top_anotador['GP']:.1f} PPG)")

print(f"\n EFICIENCIA:")
print(f"   • {top_eficiencia['PLAYER']} tiene la mayor eficiencia ({top_eficiencia['EFF']})")

print(f"\n CREACIÓN DE JUEGO:")
print(f"   • {top_asistencias['PLAYER']}: {top_asistencias['AST']} asistencias ({top_asistencias['AST']/top_asistencias['GP']:.1f} APG)")

print(f"\n DEFENSA:")
print(f"   • {top_tapones['PLAYER']} lidera en tapones con {top_tapones['BLK']}")
print(f"   • {top_robos['PLAYER']} lidera en robos con {top_robos['STL']}")

print(f"\n ANÁLISIS POR EQUIPO:")
print(f"   • {stats_equipo.iloc[0]['EQUIPO']} tiene el total de puntos más alto: {stats_equipo.iloc[0]['TOTAL_PTS']}")
print(f"   • {nba['TEAM'].nunique()} equipos representados en la liga")

print("\n══════════════════════════════════════════════════════")
```
